```python
import os
import copy
import numpy as np
import pandas as pd
import scipy
import seaborn as sns
import matplotlib.pyplot as plt 

import matplotlib.gridspec as gridspec

from IPython.display import Markdown as md
```

```python
os.chdir('/Users/chenxizhou/Dropbox/code_package/definexpfam')
true_data = np.load('data/geyser.npy').astype(np.float64)
df = copy.deepcopy(true_data[:, 0]).reshape(-1, 1)

# original data with 108.0 removed
data_waiting = (df[df != 108.0]).reshape(-1, 1)
contam_data_list = np.arange(2., 410., 4)
```

```python
bw = 9.0
landmarks1 = np.arange(42., 101., 2).reshape(-1, 1)
landmarks_name1 = 'np.arange(42., 101., 2).reshape(-1, 1)'

landmarks2 = np.arange(1., 411., 2).reshape(-1, 1)
landmarks_name2 = 'np.arange(1., 411., 2).reshape(-1, 1)'

basisfunction_name = 'Gaussian'
contam_weight = 0.01
plot_domain = (1., 410.)
plot_pts_cnt = 3000
abstol = 0.05
stepsize = 0.4
seed = 0

save_dir_ml1 = (f'data/FinExpFam-ML-basisn={len(landmarks1)}-landmarks={landmarks_name1}-' + 
                f'bw={bw}-basisfunction={basisfunction_name}-contamweight={contam_weight}-' + 
                f'plotdomain={plot_domain}-plotcnts={plot_pts_cnt}-abstol={abstol}-stepsize={stepsize}-seed={seed}')

save_dir_ml2 = (f'data/FinExpFam-ML-basisn={len(landmarks2)}-landmarks={landmarks_name2}-' + 
                f'bw={bw}-basisfunction={basisfunction_name}-contamweight={contam_weight}-' + 
                f'plotdomain={plot_domain}-plotcnts={plot_pts_cnt}-abstol={abstol}-stepsize={stepsize}-seed={seed}')


save_dir_sm = (f'data/FinExpFam-SM-basisn={len(landmarks)}-landmarks={landmarks_name}-' + 
               f'bw={bw}-basisfunction={basisfunction_name}-contamweight={contam_weight}-' + 
               f'plotdomain={plot_domain}-plotcnts={plot_pts_cnt}')
```

```python
supnorm_dict = {'bw': [bw] * len(contam_data_list), 
                'contam_data': contam_data_list.flatten(), 
                'infsup_ml1': [], 
                'infsup_ml2': []
                # 'infsup_sm': []
               }

for contam_data in contam_data_list.reshape(-1, 1):     
    
    # read in the ML influence function values 
    inf_fun_ml1 = np.load(save_dir_ml1 + f'/contam_data={contam_data}-IF-logden-newdata.npy')
    inf_fun_ml2 = np.load(save_dir_ml2 + f'/contam_data={contam_data}-IF-logden-newdata.npy')
    # inf_fun_sm = np.load(save_dir_sm + f'/contam_data={contam_data}-IF-logden-newdata.npy')
    
    supnorm_dict['infsup_ml1'].append(np.max(np.abs(inf_fun_ml1)))
    supnorm_dict['infsup_ml2'].append(np.max(np.abs(inf_fun_ml2)))
    # supnorm_dict['infsup_sm'].append(np.max(np.abs(inf_fun_sm)))    
```

```python
output_pd = pd.DataFrame(supnorm_dict)
# output_pd.to_csv(f'data/FinExpFam-SM-basisn={len(landmarks)}-landmarks={landmarks_name}-' + 
#                f'bw={bw}-basisfunction={basisfunction_name}-contamweight={contam_weight}-' + 
#                f'plotdomain={plot_domain}-plotcnts={plot_pts_cnt}-sup-norm.csv')
```
